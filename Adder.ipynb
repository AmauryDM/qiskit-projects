```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
circuit = QuantumCircuit(5, 2)
circuit.x(0)
circuit.x(1)
circuit.x(2)
circuit.barrier()
circuit.draw('mpl')
```

```python
circuit.cx(0, 3)
circuit.cx(1, 3)
circuit.cx(2, 3)
circuit.ccx(0, 1, 4)
circuit.ccx(1, 2, 4)
circuit.ccx(2, 0, 4)
circuit.barrier()
circuit.draw('mpl')
```

```python
circuit.measure(3, 0)
circuit.measure(4, 1)
circuit.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1024).result()
counts = result.get_counts()
plot_histogram(counts)
```
